## 2. Create the simData

### 3b). put these 600 point, along with their color value Y in a data frame called simData.

```python
# create the label of red and blue
label_red = np.array(['red' for i in range(300)])
label_blue = np.array(['blue' for i in range(300)])
# len(label_blue) --> 300
type(label_blue)
```

```python
label_blue
```

```python
# insert the blue and red labels into array red and blue
red_labeled = np.row_stack((red,label_red))
# red_labeled
blue_labeled = np.row_stack((blue,label_blue))
blue_labeled.shape
```

```python
# combine the red and blue array to a dataFrame called simData
temp = np.column_stack((red_labeled,blue_labeled))
# display(temp,temp.shape)
simData = pd.DataFrame(temp, index = ['X1','X2','Y'])
simData
```

```python
# row and column exchage
simData = pd.DataFrame(simData.values.T, index=simData.columns, 
                       columns=simData.index)
simData
```

```python
# sampling the simData
simData.sample(600).reset_index(drop=True)
```
